```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt
import random
import time

class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = {}

    def add_node(self, value):
        self.nodes.add(value)

    def add_edge(self, from_node, to_node, weight):
        self.edges.setdefault(from_node, []).append((to_node, weight))
        self.edges.setdefault(to_node, []).append((from_node, weight))


def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph.nodes}
    distances[start] = 0
    priority_queue = [(0, start)]
    visited = set()

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_node in visited:
            continue
        visited.add(current_node)

        for neighbor, weight in graph.edges.get(current_node, []):
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances


def bellman_ford(graph, start):
    distances = {node: float('infinity') for node in graph.nodes}
    distances[start] = 0

    for _ in range(len(graph.nodes) - 1):
        for node in graph.nodes:
            if node in graph.edges:
                for neighbor, weight in graph.edges[node]:
                    if distances[node] + weight < distances[neighbor]:
                        distances[neighbor] = distances[node] + weight

    # Check for negative weight cycles
    for node in graph.nodes:
        if node in graph.edges:
            for neighbor, weight in graph.edges[node]:
                if distances[node] + weight < distances[neighbor]:
                    raise ValueError("Graph contains negative weight cycle")

    return distances


def visualize_graph(road_network):
    G = nx.Graph()
    for node, edges in road_network.edges.items():
        for edge in edges:
            G.add_edge(node, edge[0], weight=edge[1])

    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=15)
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)
    plt.title('Road Network')
    plt.show()


def generate_random_emergency_calls(graph, num_calls):
    nodes_list = list(graph.nodes)
    return [(random.choice(nodes_list), random.choice(nodes_list)) for _ in range(num_calls)]


def main():
    # Creating a sample graph representing the road network
    road_network = Graph()
    road_network.add_node('Redmond Town Center')
    road_network.add_node('Bellevue')
    road_network.add_node('Kirkland')
    road_network.add_node('Bothell')

    road_network.add_edge('Redmond Town Center', 'Bellevue', 5)
    road_network.add_edge('Redmond Town Center', 'Kirkland', 7)
    road_network.add_edge('Bellevue', 'Kirkland', 3)
    road_network.add_edge('Bellevue', 'Bothell', 4)
    road_network.add_edge('Kirkland', 'Bothell', 6)

    # Visualize the road network graph
    visualize_graph(road_network)

    # Generate random emergency calls with a higher volume
    num_calls = 10  # Increase the number of emergency calls
    emergency_calls = generate_random_emergency_calls(road_network, num_calls)

    # Compare performance of Dijkstra's and Bellman-Ford algorithms
    for i, (start_node, destination_node) in enumerate(emergency_calls, start=1):
        print(f"\nEmergency call {i} from {start_node} to {destination_node}")

        # Measure performance of Dijkstra's algorithm
        start_time = time.time()
        shortest_path_dijkstra = dijkstra(road_network, start_node)
        dijkstra_time = time.time() - start_time
        print("Shortest path using Dijkstra's algorithm:", shortest_path_dijkstra[destination_node])
        print("Time taken by Dijkstra's algorithm:", dijkstra_time)

        # Measure performance of Bellman-Ford algorithm
        start_time = time.time()
        shortest_path_bellman_ford = bellman_ford(road_network, start_node)
        bellman_ford_time = time.time() - start_time
        print("Shortest path using Bellman-Ford algorithm:", shortest_path_bellman_ford[destination_node])
        print("Time taken by Bellman-Ford algorithm:", bellman_ford_time)

if __name__ == "__main__":
    main()
```

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt
import random
import time

class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = {}

    def add_node(self, value):
        self.nodes.add(value)

    def add_edge(self, from_node, to_node, weight):
        self.edges.setdefault(from_node, []).append((to_node, weight))
        self.edges.setdefault(to_node, []).append((from_node, weight))


def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph.nodes}
    distances[start] = 0
    priority_queue = [(0, start)]
    visited = set()

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_node in visited:
            continue
        visited.add(current_node)

        for neighbor, weight in graph.edges.get(current_node, []):
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances


def bellman_ford(graph, start):
    distances = {node: float('infinity') for node in graph.nodes}
    distances[start] = 0

    for _ in range(len(graph.nodes) - 1):
        for node in graph.nodes:
            if node in graph.edges:
                for neighbor, weight in graph.edges[node]:
                    if distances[node] + weight < distances[neighbor]:
                        distances[neighbor] = distances[node] + weight

    # Check for negative weight cycles
    for node in graph.nodes:
        if node in graph.edges:
            for neighbor, weight in graph.edges[node]:
                if distances[node] + weight < distances[neighbor]:
                    raise ValueError("Graph contains negative weight cycle")

    return distances


def visualize_graph(road_network):
    G = nx.Graph()
    for node, edges in road_network.edges.items():
        for edge in edges:
            G.add_edge(node, edge[0], weight=edge[1])

    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=15)
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)
    plt.title('Road Network')
    plt.show()


def generate_random_emergency_calls(graph, num_calls):
    nodes_list = list(graph.nodes)
    return [(random.choice(nodes_list), random.choice(nodes_list)) for _ in range(num_calls)]


def main():
    # Creating a sample graph representing the road network
    road_network = Graph()
    road_network.add_node('Redmond Town Center')
    road_network.add_node('Bellevue')
    road_network.add_node('Kirkland')
    road_network.add_node('Bothell')

    road_network.add_edge('Redmond Town Center', 'Bellevue', 5)
    road_network.add_edge('Redmond Town Center', 'Kirkland', 7)
    road_network.add_edge('Bellevue', 'Kirkland', 3)
    road_network.add_edge('Bellevue', 'Bothell', 4)
    road_network.add_edge('Kirkland', 'Bothell', 6)

    # Visualize the road network graph
    visualize_graph(road_network)

    num_calls = 900  # emergency calls
    emergency_calls = generate_random_emergency_calls(road_network, num_calls)

    # Compare performance of Dijkstra's and Bellman-Ford algorithms
    dijkstra_total_time = 0
    bellman_ford_total_time = 0

    for start_node, destination_node in emergency_calls:
        # Measure performance of Dijkstra's algorithm
        start_time = time.time()
        shortest_path_dijkstra = dijkstra(road_network, start_node)
        dijkstra_time = time.time() - start_time
        dijkstra_total_time += dijkstra_time

        # Measure performance of Bellman-Ford algorithm
        start_time = time.time()
        shortest_path_bellman_ford = bellman_ford(road_network, start_node)
        bellman_ford_time = time.time() - start_time
        bellman_ford_total_time += bellman_ford_time

    # Calculate average times
    avg_dijkstra_time = dijkstra_total_time / num_calls
    avg_bellman_ford_time = bellman_ford_total_time / num_calls

    print("Average time taken by Dijkstra's algorithm:", avg_dijkstra_time)
    print("Average time taken by Bellman-Ford algorithm:", avg_bellman_ford_time)


if __name__ == "__main__":
    main()
```
