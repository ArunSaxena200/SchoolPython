```python
import time
import random
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def bidirectional_bubble_sort(arr):
    n = len(arr)
    sorted_flag = False
    while not sorted_flag:
        sorted_flag = True
        for i in range(1, n):
            if arr[i - 1] > arr[i]:
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                sorted_flag = False
        if sorted_flag:
            break
        sorted_flag = True
        for i in range(n - 1, 0, -1):
            if arr[i - 1] > arr[i]:
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                sorted_flag = False

# Measure time taken for sorting using standard bubble sort
bubble_sort_times = []
# Measure time taken for sorting using bidirectional bubble sort
bidirectional_bubble_sort_times = []

# Define range of list sizes
list_sizes = [100, 500, 1000, 2000, 5000]

for size in list_sizes:
    arr = [random.randint(0, 10000) for _ in range(size)]
    
    start_time = time.time()
    bubble_sort(arr.copy())
    bubble_sort_times.append(time.time() - start_time)
    
    start_time = time.time()
    bidirectional_bubble_sort(arr.copy())
    bidirectional_bubble_sort_times.append(time.time() - start_time)

# Plotting
plt.plot(list_sizes, bubble_sort_times, label='Bubble Sort')
plt.plot(list_sizes, bidirectional_bubble_sort_times, label='Bidirectional Bubble Sort')
plt.xlabel('List Size')
plt.ylabel('Time (s)')
plt.title('Performance Comparison: Bubble Sort vs. Bidirectional Bubble Sort')
plt.legend()
plt.grid(True)
plt.show()
```
