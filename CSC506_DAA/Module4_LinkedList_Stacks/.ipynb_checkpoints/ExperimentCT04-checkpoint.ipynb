```python
class listNode:
    def __init__(self,value):
        self.value = value
        self.next = None
```

```python
class LL_Stack:
    def __init__(self):
        self.head = None
    def push(self,value):
        new_node = listNode(value)
        new_node.next = self.head
        self.head = new_node
    def pop(self):
        if not self.head:
            return None
        value = self.head.value
        self.head =self.head.next
        return value       
        
        
```

```python
class LL_queue:
    def __init__(self):
        self.head = None
        self.tail = None
    def enqueue(self,value):
        new_node = listNode(value)
        if not self.head:
            self.head =self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
    def dequeue(self):
        if not self.head:
            return None
        value = self.head.value
        if self.head.value == self.tail:
            self.head =self.tail =None
        else:
            self.head = self.head.next
        return value
        
        
        
```

```python
import time
import random
from collections import deque
def list_stack_performance(data):
    start_time = time.time()
    stack = []
    for item in data:
        stack.append(item)
    while stack:
        stack.pop()
    return time.time() - start_time

def list_queue_performance(data):
    start_time = time.time()
    queue = deque()
    for item in data:
        queue.append(item)
    while queue:
        queue.popleft()
    return time.time() - start_time

def linked_list_stack_performance(data):
    start_time = time.time()
    stack = LL_Stack()
    for item in data:
        stack.push(item)
    while stack.pop():
        pass
    return time.time() - start_time

def linked_list_queue_performance(data):
    start_time = time.time()
    queue = LL_queue()
    for item in data:
        queue.enqueue(item)
    while queue.dequeue():
        pass
    return time.time() - start_time

def generate_data(size):
    return [random.randint(0, 100000) for _ in range(size)]

def run_experiment(data_sizes):
    for size in data_sizes:
        data = generate_data(size)
        list_stack_time = list_stack_performance(data)
        list_queue_time = list_queue_performance(data)
        linked_list_stack_time = linked_list_stack_performance(data)
        linked_list_queue_time = linked_list_queue_performance(data)
        print(f"Data size: {size}")
        print(f"List Stack Time: {list_stack_time} seconds")
        print(f"List Queue Time: {list_queue_time} seconds")
        print(f"Linked List Stack Time: {linked_list_stack_time} seconds")
        print(f"Linked List Queue Time: {linked_list_queue_time} seconds")
        print("--------------------")

if __name__ == "__main__":
    data_sizes = [1000, 10000, 100000]
    run_experiment(data_sizes)
```
