```python
import cv2, numpy as np
import matplotlib.pyplot as plt
```

```python
image_bw = cv2.imread('detect_blob.png',0)
height, width = image_bw.shape[0:2]
```

```python
plt.imshow(image_bw)
```

```python
# manual slow method
binary = np.zeros([height,width,1],'uint8')
thresh = 85

for row in range(0,height):
    for col in range(0,width):
        if image_bw[row][col]>thresh:
            binary[row][col]=255

plt.imshow(binary)
```

```python
## Binary segmentation

ret, thresh = cv2.threshold(image_bw,thresh,255,cv2.THRESH_BINARY)
```

```python
print(ret,thresh)
```
