```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_face =  cv2.imread('faces.jpeg',1)
```

```python
plt.imshow(image_face)
```

```python
image_hsv = cv2.cvtColor(image_face,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_hsv)
```

```python
h = image_hsv[:,:,0]
s = image_hsv[:,:,1]
v = image_hsv[:,:,2]

# Show h,s,v image split by numpy horizontly (axis =1 shows it split horizontally)
hsv_split = np.concatenate((h,s,v), axis =1)
```

```python
plt.imshow(hsv_split)
```

```python
ret, min_sat = cv2.threshold(s,40,255,cv2.THRESH_BINARY) # everything in saturation channel with a value of 40 or higher will go white
```

```python
plt.imshow(min_sat)
```

```python
ret, max_hue = cv2.threshold(h,15,255,cv2.THRESH_BINARY_INV) ## INV =  inverse or normnal threshholding white become black and vice versa
```

```python
plt.imshow(max_hue)
```

```python
final = cv2.bitwise_and(min_sat,max_hue)
```

```python
plt.imshow(image_face)
```

```python
plt.imshow(final)
```
