```python
import cv2, numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('puppy.jpg',1)
```

```python
plt.imshow(image)
```

```python
image_half = cv2.resize(image,(0,0),fx =0.5,fy = 0.5)
image_strech = cv2.resize(image,(600,600))
image_near = cv2.resize(image,(600,600),interpolation  = cv2.INTER_NEAREST)
```

```python
plt.imshow(image_half)
```

```python
plt.imshow(image_strech)
```

```python
plt.imshow(image_near)
```

```python
##Rotation
M = cv2.getRotationMatrix2D((1,0),-30,1)
rotated = cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
```

```python
plt.imshow(rotated)
```
