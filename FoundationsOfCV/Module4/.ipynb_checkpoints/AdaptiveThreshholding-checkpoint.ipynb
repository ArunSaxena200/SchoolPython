```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_sudoku = cv2.imread('sudoku.png',0)
```

```python
plt.imshow(image_sudoku)
```

```python
ret, thresh_basic = cv2.threshold(image_sudoku,70,255,cv2.THRESH_BINARY)
```

```python
## missing bloacks, need to apply adaptive threshholding
plt.imshow(thresh_basic)
```

```python
thresh_adapt = cv2.adaptiveThreshold(image_sudoku,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
```

```python
plt.imshow(thresh_adapt)
##now the image is much more useful
```
