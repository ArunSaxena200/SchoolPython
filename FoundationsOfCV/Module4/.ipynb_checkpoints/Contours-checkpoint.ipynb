```python
import cv2,numpy as np, matplotlib.pyplot as plt
```

```python
image_bw = cv2.imread('detect_blob.png',1)
```

```python
plt.imshow(image_bw)
```

```python
gray = cv2.cvtColor(image_bw,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray)
```

```python
thresh = cv2.adaptiveThreshold(gray,255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
```

```python
plt.imshow(thresh)
```

```python
contours,heirarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
image2 = image_bw.copy()
```

```python
index = -1
thickness = 4
color = (255,0,255)
```

```python
cv2.drawContours(image2,contours, index, color, thickness)
```

```python
plt.imshow(image2)
```

```python
objects = np.zeros([image_bw.shape[0],image_bw.shape[1],3], 'uint8')
```

```python
for c in contours:
    cv2.drawContours(objects, [c], -1, color, -1) # c - loop array, all 
    area = cv2.contourArea(c)
    parameter = cv2.arcLength(c, True) # true indicates that arc is a closed loop as c is always closed
    M = cv2.moments(c)
    cx = int(M['m10']/M['m00'])  ## Centroid to put the center point dot
    cy = int(M['m01']/M['m00'])
    cv2.circle(objects,(cx,cy),4,(0,0,255),-1)
    print(area,parameter)
plt.imshow(objects)
```
