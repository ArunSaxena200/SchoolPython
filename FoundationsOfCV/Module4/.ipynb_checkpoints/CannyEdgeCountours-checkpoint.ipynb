```python
import cv2, numpy as np, matplotlib.pyplot as plt
```

```python
def imRGB(image):
    irgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return irgb
```

```python
img = cv2.imread('tomatoes.jpg')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
hsv = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
```

```python
res, thresh = cv2.threshold(hsv[:,:,0],25,255,cv2.THRESH_BINARY_INV)
```

```python
thresh_rgb = cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB)
plt.imshow(thresh_rgb) # overlab objects cobined tomatoes; to solve this we will use cannyedges
```

```python
edges = cv2.Canny(img,100,200,apertureSize = 3)
```

```python
plt.imshow(imRGB(edges)) ## image shows separaet tomatoes
```

```python
edge_inv = 255 -edges
```

```python
plt.imshow(imRGB(edge_inv))
```

```python
#Lets erode the image to make the edges thickker
kernel = np.ones((3,3),'uint8')
erode = cv2.erode(edge_inv,kernel,iterations = 1)
```

```python
## combine canny and erode
canny_thresh = cv2.bitwise_and(erode,thresh)
plt.imshow(imRGB(canny_thresh))
```

```python
contours, heirarchy = cv2.findContours(canny_thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
objects = img.copy()
for c in contours:
    area = cv2.contourArea(c)
    if area < 300:
        continue
        #too small 
    print(area)
    cv2.drawContours(objects,[c],-1,(255,255,255),2)
    M = cv2.moments(c)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    cv2.circle(objects,(cx,cy),5,(255,255,0),-1)
        
```

```python
plt.imshow(imRGB(objects))
```
