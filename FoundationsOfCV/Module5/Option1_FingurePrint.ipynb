```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def display_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv2.imread('latent-fingerprint.jpg',1)
display_image(img)
```

## Define the Kernel for Morphological Operations

```python
kernel = np.ones((3, 3), 'uint8')
```

## Dialte

```python
dilated = cv2.dilate(img, kernel, iterations=1)
display_image(dilated)
```

## Erosion

```python
eroded = cv2.erode(img, kernel, iterations=1)
display_image(eroded)
```

## Opening
### Opening (Erosion Followed by Dilation)
### Opening removes small noise by first eroding the image and then dilating it. This is particularly useful when there are small specks or spots in the fingerprint.

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
display_image(opening)
```

## Closing (Dilation Followed by Erosion)
### Closing fills small holes in the ridges by first dilating and then eroding the image. This is effective in closing small gaps in ridge patterns, which may improve matching.

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
display_image(closing)
```
