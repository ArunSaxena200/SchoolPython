```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def showImg(img):
 plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))   
```

```python


image_1 = np.ones((200, 200), dtype='uint8') * 128

cv2.rectangle(image_1, (50, 50), (100, 100), (255), -1)

image_2 = cv2.circle(image_1, (150, 150), 25, (255), -1)

plt.imshow(image_2,cmap = 'gray')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Assuming image_2 is the original grayscale image

# Canny edge detection
edges_canny = cv2.Canny(image_2, 50, 150)

# Sobel edge detection (calculating gradient magnitude)
sobel_x = cv2.Sobel(image_2, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image_2, cv2.CV_64F, 0, 1, ksize=3)
edges_sobel = cv2.magnitude(sobel_x, sobel_y)

# Laplacian edge detection
edges_laplacian = cv2.Laplacian(image_2, cv2.CV_64F)

# Display each result in separate subplots
plt.figure(figsize=(12, 4))

# Canny Edge Detection
plt.subplot(1, 3, 1)
plt.imshow(edges_canny, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

# Sobel Edge Detection
plt.subplot(1, 3, 2)
plt.imshow(edges_sobel, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

# Laplacian Edge Detection
plt.subplot(1, 3, 3)
plt.imshow(edges_laplacian, cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')

plt.show()
```

```python
# Add Gaussian noise
noise = np.random.normal(0, 10, image_2.shape).astype('uint8')
noisy_image = cv2.add(image_2, noise)
```

```python
plt.imshow(noisy_image, cmap='gray')
```
