```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def display_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
image_indoor = cv2.imread('indoorScene.jpg',0)
image_Outdoor = cv2.imread('OutDoorScene.jpg',0)
image_closeup = cv2.imread('closeUp.png',0)
```

```python
display_image(image_closeup)
```

```python
ret, thresh_basic = cv2.threshold(image_indoor,70,255,cv2.THRESH_BINARY)
ret_out, thresh_basic_out = cv2.threshold(image_Outdoor,70,255,cv2.THRESH_BINARY)
ret_close, thresh_basic_close = cv2.threshold(image_closeup,70,255,cv2.THRESH_BINARY)
```

```python
##BinaryThreshholding
display_image(thresh_basic)
```

```python
##BinaryThreshholding
display_image(thresh_basic_out)
```

```python
display_image(thresh_basic_close)
```

```python
thresh_adapt = cv2.adaptiveThreshold(image_indoor,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
thresh_adapt_outdoor = cv2.adaptiveThreshold(image_Outdoor,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
thresh_adapt_closeup = cv2.adaptiveThreshold(image_closeup,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
```

```python
display_image(thresh_adapt)
##now the image is much more useful
```

```python
display_image(thresh_adapt_outdoor)
##now the image is much more useful
```

```python
display_image(thresh_adapt_closeup)
##now the image is much more useful
```
