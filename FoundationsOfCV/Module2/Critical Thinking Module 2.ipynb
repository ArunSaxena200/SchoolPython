```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Load Image

```python
image = cv2.imread('puppy.jpg',1)
```

## finsing image dimensions

```python
height, width, channels = image.shape
```

```python
## create an image whichis 3 times wide and fill each channel with their color channels
b,g,r = cv2.split(image)
```

```python
rgb_split = np.empty([height,width*3, 3], 'uint8')
```

```python
rgb_split[:,0:width] = cv2.merge([b,b,b])
rgb_split[:,width:width*2] = cv2.merge([g,g,g])
rgb_split[:,width*2:width*3] = cv2.merge([r,r,r])
```

```python
plt.imshow(rgb_split)
```

```python
merged_image = cv2.merge([b,g,r])
```

```python
plt.imshow(merged_image)
```

```python
swapped_image = cv2.merge([b, r, g])
```

```python
plt.imshow(swapped_image)
```
