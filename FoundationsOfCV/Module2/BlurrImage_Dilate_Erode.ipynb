```python
import cv2, numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('puppy.jpg',1)
```

```python
plt.imshow(image)
```

```python
blur = cv2.GaussianBlur(image,(1,11),0)
```

```python
plt.imshow(blur)
```

```python
##dilate and erode -  dilate adds white abd erode add black pixels using a 5x5 matric called kernel (consider this matrix as scrubber)
kernel = np.ones((5,5),'uint8')
```

```python
dilate = cv2.dilate(image,kernel,iterations = 1)
```

```python
erode = cv2.erode(image,kernel,iterations =1)
```

```python
plt.imshow(dilate)
```

```python
plt.imshow(erode)
```
