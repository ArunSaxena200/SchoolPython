## Gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.show()
```

## Histogram matching - Best

```python
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

## Edge Detection by finding median values for gray scale

```python
median_val = np.median(gray_image)

lower_threshold = int(max(0, 0.66 * median_val))
upper_threshold = int(min(255, 1.33 * median_val))

edges = cv2.Canny(gray_image, lower_threshold, upper_threshold)
```

```python
plt.imshow(edges, cmap='gray')
plt.title(f'Lower: {lower}, Upper: {upper}')
plt.show()
```

## Morphing - Morphological operations can help in enhancing certain features, such as removing noise or emphasizing the texture.

```python
kernel = np.ones((5, 5), np.uint8)

# Apply dilation
dilated = cv2.dilate(edges, kernel, iterations=1)

# Display the dilated image
plt.imshow(dilated, cmap='gray')
plt.show()
```

## Soble edge

```python
# Load the image
image = cv2.imread('notes.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Sobel edge detection in the x-direction (horizontal edges)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # 1st derivative in x, no derivative in y

# Apply Sobel edge detection in the y-direction (vertical edges)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # No derivative in x, 1st derivative in y

# Compute the gradient magnitude
sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Convert to uint8 (standard image type for visualization)
sobel_magnitude = cv2.convertScaleAbs(sobel_magnitude)

# Display the Sobel results
plt.subplot(1, 3, 1)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')

plt.subplot(1, 3, 2)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')

plt.subplot(1, 3, 3)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Magnitude')

plt.show()
```

## Grayscale: This does not use a matrix but rather reduces the image to a single intensity channel.

## Edge Detection (Canny): The Canny edge detector involves applying two thresholds to the gradient image to detect strong and weak edges.

## Morphological Operations: The kernel matrix is used in dilation and erosion to modify the shape of objects in the image.
